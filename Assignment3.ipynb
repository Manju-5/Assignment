4. What do you think about the scoring variation for math, reading and writing based on
● Gender
● Test preparation course

```python
from statistics import variance
s1=(32829/518,37602/518,37538/518)
s2=(33172/482,31593/482,30581/482)
print(variance(s1))
print(variance(s2))
```

```python
from statistics import variance
s3=(25048/358,26555/358,26737/358)
s4=(40953/642,42640/642,41382/642)
print(variance(s3))
print(variance(s4))
```

Hence scoring variation for math, reading and writing based on gender is more for female.
Scoring variation for math, reading and writing based on test preparation course is more for those who completed the course.
